```python
#Este estudo apresenta as possibilidades que a biblioteca Matplotlib oferece para dataviz de um dataset do enem.

#É uma biblioteca que pode produzir resultados visualmente agradáveis e informativos.

#Referências:
#https://www.kaggle.com/pertence/enem-enade
#http://inep.gov.br/microdados
#https://bokeh.org/
#https://matplotlib.org/3.1.0/gallery/subplots_axes_and_figures/subplots_demo.html
#https://colab.research.google.com/github/jakevdp/PythonDataScienceHandbook/blob/master/notebooks/04.11-Settings-and-Stylesheets.ipynb#scrollTo=0hc0fYvwhEuc

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
dados = pd.read_csv('2016.csv')
```

```python
#A primeira linha do código faz um estilo aos plots, 
#que será aplicado a partir de agora automaticamente. 
#Para listar todos os estilos de plot diponiveis, use
#print(plt.style.available)

#as opcoes sao:
#'Solarize_Light2', '_classic_test_patch', 
#'bmh', 'classic', 'dark_background', 'fast', 'fivethirtyeight', 'ggplot', 'grayscale', 
#'seaborn', 'seaborn-bright', 'seaborn-colorblind', 'seaborn-dark', 'seaborn-dark-palette', 
#'seaborn-darkgrid', 'seaborn-deep', 'seaborn-muted', 'seaborn-notebook', 'seaborn-paper', 
#'seaborn-pastel', 'seaborn-poster', 'seaborn-talk', 'seaborn-ticks', 'seaborn-white', #
#'seaborn-whitegrid', 'tableau-colorblind10'

plt.style.use("dark_background")

# Usei plt.figure para estabelecer o tamanho da figura exibida na tela. Os valores (10, 7) ficaram
#boas na minha tela, mas voce pode variar e regular o melhor tamanho.

plt.figure(figsize = (10, 7))

#Para de fato criar o plot, utilizamos o método disponível no próprio Pandas para criar 
#histogramas.
dados["enem_nt_mt"].hist(bins = 40, edgecolor = "k", alpha = .6, color = "red")

#Com relação aos parâmetros, "bins" é o número de intervalos que o histograma considera para 
#criar as barras, podendo ser alterado conforme ache melhor.

#O parâmetro "edgecolor" (cor das bordas das barras) e "k" é 
#abreviação de "black", assim dando uma borda preta a cada barra do plot.

#O "alpha" é um valor de transparência das barras que adotei como 0.4 
#(altere como achar melhor) para podermos ver a grade de referência e termos 
#uma maior noção do valor a que cada barra corresponde.

#Por fim, "color" é simplesmente a cor das barras.

#Adicionamos um título ao plot e aos eixos x e y para deixar o gráfico mais 
#compreensível e completo.

plt.title("Distribuição das notas em matemática")
plt.xlabel("Notas")



#Seguindo em frente, podemos fazer uso da seleção condicional da biblioteca 
#Pandas para "darmos um zoom" nas notas em matemática acima de 900. 
#O código é relativamente o mesmo no que diz respeito ao plot.
plt.ylabel("Contagem")
```

```python
#O método value_counts() nos retorna as variaveis repetidas da coluna de uma tabela. 
# Assim, podemos selecionar as contagens e os anos com ".values" e ".index" respectivamente. 
#Armazenamos, então, em variáveis para facilitar o uso.
count = dados["nu_ano_inicio_graduacao"].value_counts().values
anos = dados["nu_ano_inicio_graduacao"].value_counts().index

plt.figure(figsize = (20, 7))

#Com "plt.subplot", conseguimos colocar mais de um plot em uma mesma figura. 
#Os parâmetros (1, 2, 1) do primeiro subplot significa que estamos criando uma figura com 
#1 linha e 2 colunas e selecionando o plot número 1, que é criado nas linhas seguintes. 
#Assim que terminamos de personalizar o primeiro plot, selecionamos o segundo plot com 
#"plt.subplot(1, 2, 2)".

#O primeiro plot é um gráfico de barras com os valores que armazenamos nas variáveis das 
#primeiras duas linhas do código. Os outros parâmetros (ec, alpha e color) 
#funcionam exatamente da mesma forma dos primeiros plots que vimos, 
#assim como os títulos do gráfico e eixos.

plt.subplot(1, 2, 1)
plt.bar(anos, count, ec = "k", alpha = .6, color = "royalblue")
plt.xlabel("Ano")
plt.title("Ano de início de graduação dos alunos que realizaram o ENADE em 2016",
          fontdict={
                    'color' : 'red',
                    'weight': 'bold',
                    'size': 14})

#O segundo plot exige um pouco mais de código. É um gráfico de pizza (plt.pie) 
#que recebe os valores das contagens da primeira linha do código. 
#Os demais parâmetros significam:
#- Labels: lista com as etiquetas de cada porção da pizza. 
#Aqui fazemos uso da variável "anos", que já está na ordem 
#correta vindo do value_counts(), tornando ela uma lista. 
#Note que se usássemos uma lista com uma ordem diferente, o resultado apresentaria 
#dados errados, não batendo cada ano com sua contagem. Outro detalhe é que o tamanho 
#da lista deve bater com o número de dados.
#- Colors: lista com cores (rgb/rgba/hex/colorname). A ordem aqui importa. O tamanho deve ser igual ao número de dados.
#- Labeldistance: distância que as etiquetas em Labels ficam. Aconselho a brincar um pouco com esse valor para ter uma noção melhor da distância que deseja.
#- Explode: lista com valores correspondentes ao quanto você quer que cada pedaço saia do círculo. É com esse parâmetro que conseguimos explodir a pizza e criar o efeito visto acima. Para os 3 primeiros consideramos 0 e para os demais aumentamos progressivamente o valor. Se quisessemos destacar apenas um ano específico, colocaríamos 0 para todos menos esse ano. O tamanho da lista deve ser o mesmo do número de dados.
#- Wedgeprops: recebe um dicionário. Poderíamos alterar diversas coisas aqui, mas usamos somente para colocar uma borda, já que "plt.pie" não recebe como parâmetro direto "edgecolor".¶
#- Textprops: dicionário para alterar de diversas formas os textos. Aqui apenas alteramos o tamanho das etiquetas.
#A linha "plt.axis("equal")" serve para garantirmos um formato redondo e não de elipse. 

plt.subplot(1, 2, 2)
plt.pie(count, 
        labels = list(anos),  
        colors = ["#20257c", "#424ad1", "#6a8ee8", "#66bbe2", "#66dee2", "#6ce2cb", "#6ad187", "#3b7f5b"],
        labeldistance = 1.1,
        explode = [0, 0, 0, .1, .2, .4, .6, .8],
        wedgeprops = {"ec": "k"}, 
        textprops = {"fontsize": 18}, 
        ) 
plt.axis("equal")
plt.title("Ano de início de graduação dos alunos que realizaram o ENADE em 2016",
         fontdict={
                    'color' : 'red',
                    'weight': 'bold',
                    'size': 18})

#Por fim, "plt.legend()" nos dá uma legenda de cara cor e etiqueta.
#Faremos um último plot para visualizarmos as notas máxima e mínima de cada área na prova do ENEM.
#Mas antes, vamos obter esses valores.

plt.legend()
```

```python
#Primeiro, ignoraremos as notas mínimas 0.
indexes = dados[dados["enem_nt_mt"] == 0].index
dados = dados.drop(indexes)
indexes = dados[dados["enem_nt_ch"] == 0].index
dados = dados.drop(indexes)
```

```python
#Então, criamos listas com os valores máximos e mínimos de cada área.
notas_max = [dados["enem_nt_cn"].max(), dados["enem_nt_ch"].max(), dados["enem_nt_lc"].max(), dados["enem_nt_mt"].max()]
notas_min = [dados["enem_nt_cn"].min(), dados["enem_nt_ch"].min(), dados["enem_nt_lc"].min(), dados["enem_nt_mt"].min()]

```

```python
#Finalmente, criamos o plot:

plt.figure(figsize = (10, 5))
bar_width = .35
index = np.arange(4)

plt.barh(index, 
         notas_max, 
         ec = "k", 
         alpha = .6, 
         color = "royalblue", 
         height = bar_width, 
         label = "Máxima")

plt.barh(index - bar_width, 
         notas_min, 
         ec = "k", 
         alpha = .6, 
         color = "darkblue", 
         height = bar_width, 
         label = "Mínima")

for i, v in enumerate(notas_max):
    plt.text(v - 50, i, str(v))
for i, v in enumerate(notas_min):
    plt.text(v - 50, i - bar_width, str(v))
        
plt.yticks(index - bar_width / 2, ("Ciências Naturais", "Ciências Humanas", "Linguagem e Códigos", "Matemática"))
plt.title("Notas máximas e mínimas em cada área (ENEM 2016)")
plt.legend()
```

```python
#Aqui criamos 2 gráficos de barras horizontais na mesma figura, obtendo como resultado o gráfico acima. O parêmtro "height" no gráfico de barras horizontais "plt.barh" indica a grossura da barra, que já havia sido designado a uma variável na segunda linha do código. "label" é o que irá aparecer na legenda para tornar o gráfico mais compreensível e completo.
#Aqui, a variável "index" é simplesmente valores de 0 a 3 para que o programa saiba onde posicionar cada barra no eixo vertical. Para o segundo "plt.barh", subtraímos desse "index" a grossura da barra, para que as barras correspondentes às notas mínimas se posicionem imediatamente abaixo das barras correspondentes às notas máximas. Caso usássemos o mesmo valor para ambos os plots, as barras das notas mínimas ficariam sobrepostas às barras das notas máximas.
#Seguindo em frente no código, os loops for são necessários para a colocação dos valores correspondentes a cada barra no topo dela. Sem isso, teríamos apenas as barras, sem os valores a que elas correspondem. Iteramos sobre as listas das notas utilizando a função "enumerate" do Python para obtermos o contador automático que ele nos fornece. Colocamos manualmente o texto correspondente aos valores com "plt.text". Sendo "i" o contador automático, neste caso de 0 a 3 (idêntico a nossa variável "index"), e "v" cada valor da lista, utilizamos "i" e "v" como posições no plot para colocação do texto. Subtraímos 50 de "v" para posicionarmos o texto um pouco dentro das barras e subtraímos "bar_width" de "i" no segundo plot para as notas mínimas devido ao nosso posicionamento das barras anteriormente. Resumindo, os dois primeiros parâmetros correspondem ao posicionamento do texto no plot. O terceiro é o texto que queremos colocar, neste caso é a forma em string de "v" (que originalmente é um float).
#"plt.yticks" nos permite personalizar o eixo vertical para exibir as áreas da prova do ENEM correspondentes às barras. "index - bar_width / 2" é uma maneira de centralizar o texto entre as duas barras, enquanto o segundo parâmetro é uma tupla com o que desejamos exibir.
```
